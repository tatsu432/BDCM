```python
# Define the structual equation
# X_i = f_i(X_{pa_i}) + U_i

def structural_eq(u, ind, x):
  if ind == 0:
    output = u
  elif ind == 1:
    output = x[0] ** 2 + u
  elif ind == 2:
    output = x[0] + u
  elif ind == 3:
    output = x[0] ** 3 + x[0] + u
  elif ind == 4:
    output = x[2] ** 2 + 0.1 + u
  elif ind == 5:
    output = x[1] * x[3] + u
  elif ind == 6:
    output = x[0] * x[4] + x[5] * x[5] + x[5] * x[0] + u
  return normalize(output)

SCM3(structural_eq=structural_eq, simple_or_complex = "simple")
```
